```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
df_titanic['Age'] = df_titanic[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(df_titanic.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sex = pd.get_dummies(df_titanic['Sex'])
print(sex)
```

```python
sex = pd.get_dummies(df_titanic['Sex'],drop_first=True)
print(sex)
```

```python
sex = pd.get_dummies(df_titanic['Sex'],drop_first=True)
embark = pd.get_dummies(df_titanic['Embarked'],drop_first=True)
```

```python
sex.head()
```

```python
embark.head()
```

```python
df_titanic = pd.concat([df_titanic, sex,embark], axis=1)
```

```python
df_titanic.head()
```

```python
df_titanic.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
df_titanic.head()
```

```python
df_titanic.tail()
```

```python
df_titanic.drop(['PassengerId'],axis=1,inplace=True)
```

```python
df_titanic.head()
```

```python
X = df_titanic.drop('Survived',axis=1)
```

```python
y = df_titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrm = LogisticRegression()
```

```python
lrm.fit(X_train,y_train)
```

```python
predictions = lrm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```
